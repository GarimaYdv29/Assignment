```python
population=pd.read_csv("C:\\Users\\Asus\\Downloads\\world_population.csv")
population.head(10)
```

```python
population= population.dropna()
print(population)
```

```python
population.sort_values(by = '2022 Population',ascending = False)
population['Country/Territory'].head(5)
```

```python
population['GDP_per_capita'] = population['Growth Rate'] / population['2022 Population']
population[['Country/Territory', 'GDP_per_capita']]
```

```python
filtered = population[
    (population['2022 Population'] > 100_000_000) &
    (population['Area (kmÂ²)'] < 500_000)
]
print(filtered[['Country/Territory', '2022 Population', 'Area (kmÂ²)']])
```

```python
population['Density'] = population['2022 Population'] / population['Area (kmÂ²)']
sorted_density = population.sort_values(by='Density', ascending=False)
print(sorted_density[['Country/Territory', 'Density']])
```

```python
students=pd.read_csv("C:\\Users\\Asus\\Downloads\\StudentsPerformance.csv")
```

```python
average_math = students['math score'].mean()
average_reading = students['reading score'].mean()
average_writing = students['writing score'].mean()

print("Average Math Score:", round(average_math, 2))
print("Average Reading Score:", round(average_reading, 2))
print("Average Writing Score:", round(average_writing, 2))
```

```python
writing_avg_gender = students.groupby('gender')['writing score'].mean()
print("Writing score by gender:\n", writing_avg_gender)

# Determine who performed better
top_gender = writing_avg_gender.idxmax()
print(f"{top_gender.title()}s performed better in writing.")
```

```python
students['total score'] = students['math score'] + students['reading score'] + students['writing score']
students['percentage'] = students['total score'] / 3

# View first few rows with new columns
students[['math score', 'reading score', 'writing score', 'total score', 'percentage']].head()
```

```python
import matplotlib.pyplot as plt

avg_scores = students.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
avg_scores.plot(kind='bar', figsize=(10, 6))
plt.title("Average Scores by Parental Level of Education")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
high_performers = students[
    (students['math score'] >= 90) &
    (students['reading score'] >= 90) &
    (students['writing score'] >= 90)
]

print("Students scoring above 90 in all subjects:")
print(high_performers)
```
